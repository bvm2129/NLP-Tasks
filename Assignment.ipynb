```python
!pip install huggingface_hub # install required package

from huggingface_hub import InferenceClient
from getpass import getpass
import textwrap #import required modules

HF_TOKEN = getpass("Enter your Hugging Face API token: ") # get your HF token

# create an inference client
# InferenceClient - Python class from the huggingface_hub
client = InferenceClient(
    model="mistralai/Mistral-7B-Instruct-v0.2",
    token=HF_TOKEN
)

# Prepare your conversation (as a chat)
user_question = input("Enter your question: ")

messages = [
    {"role": "system", "content": "You are a helpful and knowledgeable assistant."},
    {"role": "user", "content": user_question}
]

# send the chat message
response = client.chat_completion(messages=messages, max_tokens=200)

# display the answer
print("\nAnswer:")
print(textwrap.fill(response.choices[0].message.content.strip(), width=80))
```

```python
import torch
from transformers import pipeline
# torch makes the program run faster and lighter on the PC GPU

# initializing the model and integrating it
translator = pipeline("translation", model="facebook/nllb-200-distilled-600M", torch_dtype=torch.float16)

text = """\
My puppy is adorable,  \
Your kitten is cute.
Her panda is friendly.
His llama is thoughtful. \
We all have nice pets!"""

# every language has it's code as per taken model...
# "https://github.com/facebookresearch/flores/blob/main/flores200/README.md#languages-in-flores-200" **language reference**

# we are trying the hindi language
text_translated = translator(text, src_lang="eng_Latn", tgt_lang="hin_Deva")
print(text_translated)
```

```python
# now we are initializing another bot for question answering
# initializing the model and integrating it
qa_bot=pipeline("question-answering", model="deepset/roberta-base-squad2")

# we are asking the question for it to answer from the provided text after analyzing it
question="What is outer-space?"
text="""
Outer space, or simply space, is the expanse that exists beyond Earth's atmosphere
and between celestial bodies.[1] It contains ultra-low levels of particle densities,
constituting a near-perfect vacuum[2] of predominantly hydrogen and helium plasma,
permeated by electromagnetic radiation, cosmic rays, neutrinos, magnetic fields and
dust. The baseline temperature of outer space, as set by the background radiation
from the Big Bang, is 2.7 kelvins (−270 °C; −455 °F).
"""

output=qa_bot(question=question, context=text)
print(output)
```

```python
# now, we are checking if the bot can understand the emotion we are in
# for this, we don't particularly need to initialize a model for this (this will indirectly be taken care of huggingface)
sentiment=pipeline("sentiment-analysis")

opinion1=sentiment("The movie was fantastic!")
# checking it with a happy response
print(opinion1)

opinion2=sentiment("I hated the game, it was terrible.")
# checking it with a sad or hated response
print(opinion2)
```

```python
# upto now, we looked at models that directly took a single line command and gave outputs (zero-shot prompting)
# but now, we will give certain examples to the model, and then provide the question (few-shot prompting)
from transformers import pipeline

# Initializing the model and integrating it
translation =pipeline("translation_en_to_es",model="google/flan-t5-large")

# giving the examples and question combined, in the prompt
prompt="""Translate to spanish:
English: Good morning
Spanish: buen día
English: What are you doing?
Spanish:
give me only answer.
"""

print(translation(prompt))
```
